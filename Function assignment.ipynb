##  6. Explain the concept of generators in Python and how they are defined.

### Ans 6 :

Generators are a type of iterator that produce values one at a time,  on demand , using the yield keyword. 
Unlike normal functions that return all values at once, generators yield values one by one, saving memory.
Yield is creating generator or iterator object. 

```python
def square_number_generators(n):
    for i in range(n):
        yield i ** 2
square_number_generators(6)
```

```python
gen = square_number_generators(6)
next(gen)
```

```python
next(gen)
```

##  7. What are the advantages of using generators over regular functions?

### Ans 7 :

Generators only produce one value at a time, so they are much more memory efficient mainly for large datasets. 

Generators are useful for generating infinite sequences like Fibonacci numbers, which is impossible with normal functions that return lists.

```python
def square_num(n):
    result = []
    for i in n:
        result.append(i ** 2)
    return result
square_num([5,6,7,8,9])
```

```python
def fib(n):
    a = 0
    b = 1
    for i in range(n):
        yield a 
        a , b = b , a+b
```

```python
f = fib(1000)
f
```

```python
next(f)
```

```python
next(f)
```

##  10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

### Ans 10 :

map() applies a function to all items in an iterable and returns a new iterable. 
reduce()  applies a function to the items of an iterable, reducing them to a single value.
filter() is an iterable based on a condition , function that returns True or False.

```python
def square(b):
    return b * b
num = [1,2,3,4]
squared_num = map(square, num)
squared_num
```

```python
from functools import reduce
reduce(lambda x,y: x+y, [1,2,3])
```

```python
from functools import reduce
num = [1,2,3,4,5]
sq = map(lambda x: x**2, num)  # map: square all numbers
even = filter(lambda x: x % 2 == 0, sq)   # filter: keep only even numbers
print("squared_num:", sq)
print("even_squared_num:", even)
```

```python
sum_of_evens = reduce(lambda x,y: x+y, even)
print(sum_of_evens)
```

##  4. Write a Python function that checks if a given number is prime or not from 1 to 200.

### Ans 4 :

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Prime numbers from 1 to 200
prime_num = [n for n in range(1 , 201) if is_prime(n)]
print('Prime numbers from 1 to 200:' , prime_num)
```

##  6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

### Ans 6 :

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
        
for value in powers_of_two(5):
    print(value)
```

##  7. Implement a generator function that reads a file line by line and yields each line as a string.

### Ans 7 :

```python
def read_file_line_by_line(filepath):
    with open(filepath, 'r') as file:
        for line in file:
            yield line.strip()

# assume there is a file named 'fruit.txt'
for line in read_file_line_by_line('fruit.txt'):
    print(line)
```

##  9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

### Ans 9 :

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
celsius_temperatures = [0,20,40,50]   # list of temperatures in celsius
# convert to Fahrenheit using map()
fahrenheit_temperatures = map(celsius_to_fahrenheit, celsius_temperatures)

for temp in fahrenheit_temperatures:   # print each fahrenheit temperatures
    print(temp)
```

##  10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

### Ans 10 :

```python
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, s))

s = "Hello Everyone"
result = remove_vowels(s)
print("String without vowels:", result)
```
